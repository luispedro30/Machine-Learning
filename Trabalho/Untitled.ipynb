```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_portuguese = pd.read_csv('student-por.csv',sep=';',)
display(df_portuguese)
df_portuguese.dtypes
```

```python
df_maths = pd.read_csv('student-mat.csv',sep=';',)
display(df_maths)
df_maths.dtypes
```

```python
print(df_portuguese.isnull().sum())
print(df_maths.isnull().sum())
```

```python
pd.options.plotting.backend = "plotly"
np.random.seed(1)

df = pd.DataFrame(dict(
    Maths=df_maths['absences'],
    Portuguese=df_portuguese['absences'])
)
fig = df.boxplot(labels={
                     "value": "Absences",
                     "variable": "Subjects"},
    title = 'Boxplot of absences')
fig.show()
```

```python
pd.options.plotting.backend = "plotly"
np.random.seed(1)

df = pd.DataFrame(dict(
    Maths=df_maths['age'],
    Portuguese=df_portuguese['age'])
)
fig = df.boxplot(labels={
                     "value": "Absences",
                     "variable": "Subjects"},
    title = 'Boxplot of absences')
fig.show()
```

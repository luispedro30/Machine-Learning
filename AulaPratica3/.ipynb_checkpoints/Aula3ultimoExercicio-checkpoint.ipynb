```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import preprocessing
import sklearn.linear_model as lm
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv("../datasets/kc_house_data.csv")
df.head()
```

Verify if there any null values (Missing values)

```python
# Total missing values for each feature
print(df.isnull().sum())
```

Verify the types of the columns in order to confirm if we need some transformation

```python
df.dtypes
```

# Drop of the columns that are not necessary like Date or Zipcode

```python
df.drop('date', axis=1, inplace=True)
df.drop('zipcode',axis=1, inplace=True)
```

```python
df.head()
```

Crie um modelo de regressÃ£o linear univariada para prever o custo da casa em funÃ§Ã£o da Ã¡rea de implantaÃ§Ã£o.

```python
x_area = df['sqft_living15']
x_area = np. expand_dims(x_area, 1)
y_area = df['price']

lr = lm.LinearRegression()
lr.fit(x_area, y_area)
print('intercept : ', lr.intercept_)
print('coefficient : ', lr.coef_)
```

```python
independent_variables = ['id','bedrooms', 'bathrooms',
                         'sqft_living','sqft_lot','floors',
                         'waterfront', 'view','condition',
                         'sqft_above','sqft_basement',
                         'yr_built','yr_renovated','lat','long',
                         'sqft_living15','sqft_lot15']

dependent_variables = ['price']
```

Atualization of grade values

```python
#Atualization of grade values
df.loc[df.grade<=3,'grade'] = 1
df.loc[(df['grade'] > 3) & (df['grade'] < 7),'grade']=2
df.loc[(df['grade'] >= 7) & (df['grade'] <11),'grade']=3
df.loc[df.grade>=11,'grade'] = 4
```

Boxplot of each variable in order to check the outliers

```python
for column in df:
    plt.figure()
    df.boxplot([column],grid='true', color='blue',fontsize=10, rot=30)
plt.show()
```

```python
#Faz sentido termos meio andar?
print(df.floors.unique())
#Faz sentido termos 1.75 de casas de banho?
print(df.bathrooms.unique())
#33 quartos? Remove?
print(df.bedrooms.unique())
```

We will considere the value of 33 bedrooms as an outlier, so we will remove it

```python
print(df.shape)
df = df[df.bedrooms != 33]

print(df.bedrooms.unique())
print(df.shape)
```

Using Vif we will check if we can delete some variable

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

thresh = 10
for i in np.arange(0, len(independent_variables)):
    vif = [variance_inflation_factor(df[independent_variables].values,ix)
        for ix in range(df[independent_variables].shape[1])]
    maxloc = vif.index(max(vif))
    if max(vif) > thresh:
        print('vif :', vif)
        print('dropping', df[independent_variables].columns[maxloc],
              'at index', maxloc)
        del independent_variables[maxloc]
    else:
        break
```

```python
print(independent_variables)
```

As there are a lot of outliers, and we don't want some variables to be more
important than others, we will normalize the data

```python
X = df[independent_variables]
y = df[dependent_variables]
```

```python
X_train , X_test , y_train , y_test = train_test_split (X, y, train_size =.7,
                                                        random_state =1)
```

Data standardization

```python
scaler = preprocessing.StandardScaler().fit(X)
X = scaler.transform(X)
```

RegressÃ£o linear Multivariada

```python
lr = lm.LinearRegression()
lr.fit(X_train,y_train)
print(lr.coef_)
print(lr.intercept_)

#make predictions on the training dataset
y_train_pred = lr.predict(X_train)
print('Train MAE : ',metrics.mean_absolute_error(y_train,y_train_pred))
print('Train RMSE',np.sqrt(metrics.mean_squared_error(y_train,y_train_pred)))

#make predictions on the testing set
y_test_pred = lr.predict(X_test)
print('Train MAE : ',metrics.mean_absolute_error(y_test,y_test_pred))
print('Train RMSE',np.sqrt(metrics.mean_squared_error(y_test,y_test_pred)))
```

```python
# check our accuracy for each degree, the lower the error the better!
number_degrees = [1,2,3]
plt_mean_squared_error = []
for degree in number_degrees:
    poly_model = PolynomialFeatures(degree=degree)
    poly_x_values = poly_model.fit_transform(X)
    poly_model.fit(poly_x_values, y)
  
    regression_model = lm.LinearRegression()
    regression_model.fit(poly_x_values, y)
    
    y_pred = regression_model.predict(poly_x_values)
    plt_mean_squared_error.append(np.sqrt(metrics.mean_squared_error(y, y_pred)))

    print('R- Squared for degree ', str(degree), '=',r2_score (y, y_pred))
    print('Train MAE : ',metrics.mean_absolute_error(y, y_pred))
    print('Train RMSE',np.sqrt(metrics.mean_squared_error(y, y_pred)))
```
